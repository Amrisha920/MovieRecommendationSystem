Import Libraries

### Change Dimension

```python
X_train.ndim # To check dimension
```

```python
X_train.shape
```

```python
X_train=np.expand_dims(X_train,-1)
X_test=np.expand_dims(X_test,-1)
```

```python
X_train.ndim
```

```python
X_train.shape
```

```python
X_train=X_train/255 #To scale the data bw  0-1
X_test=X_test/255
```

```python
X_train[0]
```

### Split Dataset

```python
from sklearn.model_selection import train_test_split                                                                                        #provide supervised and unsupervised learning algoes
X_train,X_val, y_train,y_val=train_test_split(X_train,y_train,test_size=0.2,random_state=2020)                                                          #data split in training and validation
```

```python
X_train.shape,y_train.shape
```

### Convolutional Neural Network-Model Building

```python
model=keras.models.Sequential([                                                                                                                 #load models,layers inside sequential
                         keras.layers.Conv2D(filters=32, kernel_size=3 ,strides=(1,1),padding='valid',activation='relu',input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),                                                                           #2,2 matrix converted into single matrix
                         keras.layers.Flatten(),                                                                                                       #convert into single row/vector
                         keras.layers.Dense(units=128 , activation='relu'),                                                                         #create neural network,128 neurons
                         keras.layers.Dense(units=10 , activation='softmax')                                                                                     ,#probability finding


                         ])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=512, verbose=1,validation_data=(X_val,y_val))                                                                     #train the model
```

### Test the Model



```python
model.predict(np.expand_dims(X_test[0],axis=0)).round(2)                                                                                                    # predict model giving the correct value or not
```

```python
np.argmax(model.predict(np.expand_dims(X_test[0],axis=0)).round(2))                                                                                           #predict maximam value
```

```python
y_test[0]
```

```python
y_pred=model.predict(X_test).round(2)
y_pred
```

```python
model.evaluate(X_test,y_test)                                                                                               #how much accuracy is the model giving on the test dataset
```

### Visualize Output

```python
 plt.figure(figsize=(16,16))

j=1
for i in np.random.randint(0,1000,60):
  plt.subplot(10,6,j); j+=1
  plt.imshow(X_test[i], cmap="gist_stern_r")
  plt.title('Actual={} / {} \nPredicted={} / {}'.format(class_labels[y_test[i]] , y_test[i],class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
  plt.axis('off') # off the axis
```

### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

plt.figure(figsize=(16,9))
y_pred_labels=[np.argmax(label) for label in y_pred ]                                                                                       #single label  value and store in y_pred
cm = confusion_matrix(y_test,y_pred_labels)                                                                                               #the info we get will in the fiorm of matrix

#show cm
sns.heatmap(cm, annot=True,fmt='d',xticklabels=class_labels,yticklabels=class_labels)                                                                       #annot-value showm in box
```

```python
from sklearn.metrics import classification_report                                                                                                 #we will get the recall value here
cr=classification_report(y_test,y_pred_labels,target_names=class_labels)
print(cr)
```

### Convolutional Neural Network -Complex Model

```python
model=keras.models.Sequential([
                         keras.layers.Conv2D(filters=32, kernel_size=3 ,strides=(1,1),padding='valid',activation='relu',input_shape=[28,28,1]),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Conv2D(filters=64,kernel_size=3,strides=(2,2),padding='same',activation='relu'),
                         keras.layers.MaxPooling2D(pool_size=(2,2)),
                         keras.layers.Flatten(),
                         keras.layers.Dense(units=128 , activation='relu'),
                         keras.layers.Dropout(0.25),                                                                                         #model will be regulize from overfitting
                         keras.layers.Dense(units=256 , activation='relu'),
                          keras.layers.Dropout(0.25),
                          keras.layers.Dense(units=128 , activation='relu'),
                          keras.layers.Dense(units=128 , activation='softmax')



                         ])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=20, batch_size=512, verbose=1,validation_data=(X_val,y_val))
```

```python
model.evaluate(X_test,y_test)
```
